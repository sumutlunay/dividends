```python
%%stata
adopath++ "/home/kennesaw/Sunay/adofiles"

use "/home/kennesaw/Sunay/Dividends/levelfull.dta", clear

gen group1 = manag+employ
gen group2 = pe1+pe2

drop if (group1+group2==0)

replace group1=0 if missing(group1)
replace group2=0 if missing(group2)

gen public=0
replace public=1 if private==0

set seed 1984
gen u=uniform()
sort u

psmatch2 private int_cov loss intang ppe growth i.fyear i.ind3 i.atRank i.daRank, logit caliper(.001) noreplace neighbor(1)
gen pair = _id if _treated==0
replace pair = _n1 if _treated==1
bysort pair: egen paircount = count(pair)
drop if paircount !=2

areg div2 private salgrowth logat da opprofit, absorb(fyear) robust cluster(gvkey)
outreg2 using myfile,e(r2_a) stats (coef tstat) bdec(3) tdec(2) excel replace

gen in_sample=e(sample)
total private if in_sample==1
drop in_sample

areg repurch2 private salgrowth logat da opprofit, absorb(fyear) robust cluster(gvkey)
outreg2 using myfile,e(r2_a) stats (coef tstat) bdec(3) tdec(2) excel append

gen in_sample=e(sample)
total private if in_sample==1
drop in_sample

areg payout2 private salgrowth logat da opprofit, absorb(fyear) robust cluster(gvkey)
outreg2 using myfile,e(r2_a) stats (coef tstat) bdec(3) tdec(2) excel append

gen in_sample=e(sample)
total private if in_sample==1
drop in_sample

areg div2 group1 group2 salgrowth logat da opprofit, absorb(fyear) robust cluster(gvkey)
outreg2 using myfile,e(r2_a) stats (coef tstat) bdec(3) tdec(2) excel append

test group1 = group2
gen in_sample=e(sample)
total  group1 group2 if in_sample==1
drop in_sample


areg repurch2 group1 group2 salgrowth logat da opprofit, absorb(fyear) robust cluster(gvkey)
outreg2 using myfile,e(r2_a) stats (coef tstat) bdec(3) tdec(2) excel append

test group1 = group2
gen in_sample=e(sample)
total  group1 group2 if in_sample==1
drop in_sample


areg payout2 group1 group2 salgrowth logat da opprofit, absorb(fyear) robust cluster(gvkey)
outreg2 using myfile,e(r2_a) stats (coef tstat) bdec(3) tdec(2) excel append

test group1 = group2
gen in_sample=e(sample)
total  group1 group2 if in_sample==1
drop in_sample

/*Industry Fixed Effects*/
egen fe=group(fyear FFI48)

areg div2 private salgrowth logat da opprofit, absorb(fe) robust cluster(gvkey)
outreg2 using myfile,e(r2_a) stats (coef tstat) bdec(3) tdec(2) excel append

gen in_sample=e(sample)
total private if in_sample==1
drop in_sample

areg repurch2 private salgrowth logat da opprofit, absorb(fe) robust cluster(gvkey)
outreg2 using myfile,e(r2_a) stats (coef tstat) bdec(3) tdec(2) excel append

gen in_sample=e(sample)
total private if in_sample==1
drop in_sample

areg payout2 private salgrowth logat da opprofit, absorb(fe) robust cluster(gvkey)
outreg2 using myfile,e(r2_a) stats (coef tstat) bdec(3) tdec(2) excel append

gen in_sample=e(sample)
total private if in_sample==1
drop in_sample

areg div2 group1 group2 salgrowth logat da opprofit, absorb(fe) robust cluster(gvkey)
outreg2 using myfile,e(r2_a) stats (coef tstat) bdec(3) tdec(2) excel append

test group1 = group2
gen in_sample=e(sample)
total  group1 group2 if in_sample==1
drop in_sample


areg repurch2 group1 group2 salgrowth logat da opprofit, absorb(fe) robust cluster(gvkey)
outreg2 using myfile,e(r2_a) stats (coef tstat) bdec(3) tdec(2) excel append

test group1 = group2
gen in_sample=e(sample)
total  group1 group2 if in_sample==1
drop in_sample


areg payout2 group1 group2 salgrowth logat da opprofit, absorb(fe) robust cluster(gvkey)
outreg2 using myfile,e(r2_a) stats (coef tstat) bdec(3) tdec(2) excel append

test group1 = group2
gen in_sample=e(sample)
total  group1 group2 if in_sample==1
drop in_sample
```
