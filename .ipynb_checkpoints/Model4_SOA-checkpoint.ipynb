```python
%%stata
adopath++ "/home/kennesaw/Sunay/adofiles"

use "/home/kennesaw/Sunay/Dividends/levelfull.dta", clear

bysort gvkey(fyear): gen lagdpsh = dvpsx_f[_n-1]
gen deldps = dvpsx_f - lagdpsh

reg deldps lagdpsh eps

levelsof gvkey, local(firm)
foreach i of local firm {
    reg deldps lagdpsh eps if gvkey == "`i'"
    estimates store model`i'
    esttab model`i', b se r2
}
```

```python
%%stata
adopath++ "/home/kennesaw/Sunay/adofiles"

use "/home/kennesaw/Sunay/Dividends/levelfull.dta", clear

bysort gvkey(fyear): gen lagdpsh = dvpsx_f[_n-1]
gen deldps = dvpsx_f - lagdpsh

quietly by gvkey: eststo: quietly reg deldps lagdpsh eps
        esttab
```

```python
%%stata
adopath++ "/home/kennesaw/Sunay/adofiles"

clear
tempfile all_results
save `all_results', emptyok

use "/home/kennesaw/Sunay/Dividends/levelfull.dta", clear

bysort gvkey(fyear): gen lagdpsh = dvpsx_f[_n-1]
gen deldps = dvpsx_f - lagdpsh

append using `all_results'
    save "`all_results'", replace
statsby _b e(N), by(gvkey) nodots: reg deldps lagdpsh eps if private==1
list, compress noobs
```

```python
%%stata
adopath++ "/home/kennesaw/Sunay/adofiles"

clear
tempfile all_results
save `all_results', emptyok

use "/home/kennesaw/Sunay/Dividends/levelfull.dta", clear

bysort gvkey(fyear): gen lagdpsh = dvpsx_f[_n-1]
gen deldps = dvpsx_f - lagdpsh
gen public = 1
replace public = 0 if private == 1

foreach v of varlist public private {
    tempfile `v'
    statsby, by(gvkey) saving(``v''): reg deldps lagdpsh eps if `v'
}
foreach v of varlist public private {
    clear
    use ``v''
    gen event_era = "`v'"
   append using `all_results'
    save "`all_results'", replace
}
```
