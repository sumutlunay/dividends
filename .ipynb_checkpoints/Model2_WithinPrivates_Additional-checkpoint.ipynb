"Repurchase test makes more sense with annual data. Dividend test makes sense also qith quarterly data, though we may want to limit to firms-quarters paying dividends"

```python
%%stata
adopath++ "/home/kennesaw/Sunay/adofiles"

use "/home/kennesaw/Sunay/Dividends/withinprivates_q.dta", clear

gen group1 = manag+employ
gen group2 = pe1+pe2

drop if (group1+group2==0)
egen quarter=group(datacqtr)

areg div1 group1 salgrowth logat da opprofit if payer==1, absorb(quarter) robust cluster(gvkey)
outreg2 using myfile,e(r2_a) stats (coef tstat) bdec(3) tdec(2) excel replace

areg div2 group1 salgrowth logat da opprofit if payer==1, absorb(quarter) robust cluster(gvkey)
outreg2 using myfile,e(r2_a) stats (coef tstat) bdec(3) tdec(2) excel append
```

Logistic Regressions

```python
%%stata
adopath++ "/home/kennesaw/Sunay/adofiles"

use "/home/kennesaw/Sunay/Dividends/withinprivates.dta", clear

gen group1 = manag+employ
gen group2 = pe1+pe2

drop if (group1+group2==0)

gen repayer=0
replace repayer=1 if repurch>0

gen distributor=0
replace distributor=1 if payer+repayer>1

logit payer group1 salgrowth logat da opprofit i.fyear, robust cluster(gvkey)
outreg2 using myfile,e(r2_p) stats (coef tstat) bdec(3) tdec(2) excel replace

logit repayer group1 salgrowth logat da opprofit i.fyear, robust cluster(gvkey)
outreg2 using myfile,e(r2_p) stats (coef tstat) bdec(3) tdec(2) excel append

logit distributor group1 salgrowth logat da opprofit i.fyear, robust cluster(gvkey)
outreg2 using myfile,e(r2_p) stats (coef tstat) bdec(3) tdec(2) excel append


/*Quarterly*/
use "/home/kennesaw/Sunay/Dividends/withinprivates_q.dta", clear

gen group1 = manag+employ
gen group2 = pe1+pe2

drop if (group1+group2==0)
egen quarter=group(datacqtr)

gen repayer=0
replace repayer=1 if netrep>0

gen distributor=0
replace distributor=1 if payer+repayer>1

logit payer group1 salgrowth logat da opprofit i.quarter, robust cluster(gvkey)
outreg2 using myfile,e(r2_p) stats (coef tstat) bdec(3) tdec(2) excel append

logit repayer group1 salgrowth logat da opprofit i.quarter, robust cluster(gvkey)
outreg2 using myfile,e(r2_p) stats (coef tstat) bdec(3) tdec(2) excel append

logit distributor group1 salgrowth logat da opprofit i.quarter, robust cluster(gvkey)
outreg2 using myfile,e(r2_p) stats (coef tstat) bdec(3) tdec(2) excel append
```
