```python
import ipystata
from ipystata.config import config_stata
config_stata('/usr/local/bin/stata')
```

Annual Results

Dependent Variable Scaled by Assets

```python
%%stata
adopath++ "/home/kennesaw/Sunay/adofiles"

use "/home/kennesaw/Sunay/Dividends/withinprivates.dta", clear

gen group1 = manag+employ
gen group2 = pe1+pe2

drop if (group1+group2==0)

areg div2 group1 salgrowth logat da opprofit, absorb(fyear) robust cluster(gvkey)
outreg2 using myfile,e(r2_a) stats (coef tstat) bdec(3) tdec(2) excel replace

areg repurch2 group1 salgrowth logat da opprofit, absorb(fyear) robust cluster(gvkey)
outreg2 using myfile,e(r2_a) stats (coef tstat) bdec(3) tdec(2) excel append

areg payout2 group1 salgrowth logat da opprofit, absorb(fyear) robust cluster(gvkey)
outreg2 using myfile,e(r2_a) stats (coef tstat) bdec(3) tdec(2) excel append
```
